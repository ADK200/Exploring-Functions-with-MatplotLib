Цель задания - исследование функции 
$$\LARGE f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$$
при различных значениях $\alpha$ и $\beta$.

Для каждой задачи необходимо:

- Построить график (размер графика должен быть достаточным, чтобы визуально увидеть особенности изучаемых функций), график каждой функции должен быть одного цвета для одного значения $\alpha$ и $\beta$.

___________
<font color='blue'>
The purpose of the researching is to study the function
$$\LARGE f(x) = \frac{x^{\beta} + \alpha^{\beta}}{x^{\beta} }$$
with different values $\alpha$ и $\beta$.

For each task needs:

- Build a graph (the size of the graph should be sufficient to visually see the features of the studied functions), the graph of each function should be of the same color for one value of $ \ alpha $ and $ \ beta $.</font>


________________________________


Импортируем библиотеки и пишем исследуемую функцию

<font color='blue'>Import libraries and write the investigated function</font>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as  plt
%matplotlib inline

def f(x, alpha, beta):
    return (((x ** beta) + (alpha ** beta)) / (x ** beta))


np.seterr(divide='ignore', invalid='ignore')
```

# Задача №1 | <font color='blue'>Task №1</font>
## Часть 1 | <font color='blue'>Part 1</font>

Построить в общих осях графики для | <font color='blue'>Plot common axes graphs for:</font> 

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
# Задача №1 
# Часть 1
x = np.linspace(-5, 5, 105)
fig = plt.figure(figsize=(15,10))
axes  = fig.add_axes([0, 0, 1, 1])
axes.plot(x, f(x, 1, 1), "r")
axes.plot(x, f(x, 2, 1), "b")
axes.plot(x, f(x, 1, 2), "g")

plt.ylim(-20, 30)
plt.xlabel("x", fontsize = 30)
plt.ylabel("y", fontsize = 30)
axes.legend(["α = 1, β = 1", "α = 2, β = 1", "α = 1, β = 2"], fontsize = 20)
axes.set_title("График функции f(x)=" '$\\frac{x^β + α^β}{x^β}$' " при различных α и β", fontsize = 20)

plt.show()

# plt.savefig("myimage.png",  bbox_inches='tight')
```

# Часть 3 |<font color='blue'> Part 3</font>

Построить в общих осях графики для $x<0$.

<font color='blue'>Plot common axes graphs for $x<0$.</font>

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

<font color='blue'>On the same graph, make 1 inset showing the behavior of the charts as x moves away from 0 to $-\infty$.</font> 

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

<font color='blue'>Demonstrate the possibility (or impossibility) of intersections and aspiration of functions. Also plot the line f(x) = 0 on the graphs.</font>

```python
# Задача №1 
# Часть 3




x = np.linspace(0, -5, 105)

fig = plt.figure(figsize=(15,10))
axes  = fig.add_axes([0, 0, 1, 1])

plt.ylim(-10, 30)
axes.plot(x, f(x, 1, 1), "r")
axes.plot(x, f(x, 2, 1), "b")
axes.plot(x, f(x, 1, 2), "g")
axes.plot(x, [0 for i in x], "black")
axes.set_title("График функции f(x)=" '$\\frac{x^β + α^β}{x^β}$' " при различных α и β" " где x < 0" , fontsize = 20)
axes.legend(["α = 1, β = 1", "α = 2, β = 1", "α = 1, β = 2", "f(x) = 0"], fontsize = 20)
plt.ylabel("y", fontsize = 15)
plt.xlabel("x", fontsize = 15)

# Где x < 0 поведение графиков при удалении x от 0 к −∞
x2 = np.linspace(0, -10)
axes2  = fig.add_axes([0.3, 0.4, 0.45, 0.45])
axes2.plot(x2, f(x2, 1, 1), "r")
axes2.plot(x2, f(x2, 2, 1), "b")
axes2.plot(x2, f(x2, 1, 2), "g")
axes2.plot(x2, [1 for i in x2], "black")

axes2.set_title("Где x < 0 поведение графиков при удалении x от 0 к −∞")
axes2.legend(["α = 1, β = 1", "α = 2, β = 1", "α = 1, β = 2", "f(x) = 1"], fontsize = 20)
plt.ylabel("y", fontsize = 15)
plt.xlabel("x", fontsize = 15)
plt.show()
```

## Выводы | <font color='blue'>Сonclusions</font>


1) При $\beta$ = 1, графики пересекают ось абсцисс;

<font color='blue'>When $\beta$ = 1, the graphs cross the abscissa;</font>

2) При $\alpha$ = 1, графики пересекаются в точке (1, 2);

<font color='blue'>When $\alpha$ = 1, the graphs intersect at the point (1, 2);</font>

3) Функции имеют разрыв второго рода при х = 0;

<font color='blue'>The functions have a discontinuity of the second kind at x = 0;</font>

4) В случае когда $\beta$ = 1, левосторонний и правосторонний пределы равны $-\infty$ и $+\infty$

<font color='blue'>In the case where $\beta$ = 1, the left-side and right-side limits are equal to $-\infty$ and $+\infty$</font>
